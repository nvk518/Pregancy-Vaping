```python
import pandas as pd
import re

downloads_folder_path = '/content/drive/My Drive/'

xlsx_files = ['Vape_pregnant_final.xlsx', 'Vape_pregnancy_final.xlsx', 'Vaping_pregnancy_final.xlsx', 'e-cigarette_pregnant_final.xlsx', 'e-cigarette_pregnancy_final.xlsx', 'Vaping_pregnant_final.xlsx']
# Set of column names to be removed containing any strings present in this list
to_remove = set(['avatar', 'markup_text', 'is_commerce', 'Unnamed', 'user_mode', '/start', '/end', '/begin', 'allow_share', 'whatsapp_desc', 'aweme_id', 'uid', 'name_field', 'tt_now_log_status', 'is_hash_tag', 'field', 'sub_type_0', 'download_prompt_ts', 'height', 'width', 'share_desc_info', 'share_title', 'user_id', 'special_lock', 'story_status', 'search_user_name', '/type', 'sub_type', 'hashtag', 'search_user_desc', 'user_rate','tag_id', 'line_idx'])
cols = {}

def extract_hashtags(text):
    if isinstance(text, str):
        hashtags = re.findall(r'#\w+', text)
        return set(hashtags)
    else:
        print(text)
        return set([])

for xlsx_file in xlsx_files:

    xlsx_file_path = downloads_folder_path + xlsx_file

    df = pd.read_excel(xlsx_file_path, engine='openpyxl')
    # extract video_id from share_url
    df["video_id"] = df['share_info/share_url'].str.extract(r'\/video\/(\d+)\?_r=', expand=False)
    for col in df.columns:
      # drop selected columns
      for val in to_remove:
        if val in col:
          df.drop(col, axis=1, inplace=True)
          break
      # if / occurs, remove hierarchical structure of columns by renaming
      if '/' in col:
        splitted = col.split('/')
        df.rename(columns={col: splitted[-1]}, inplace=True)

    # Remove insignificant columns remaining
    df.drop("uri", axis=1, inplace=True)
    df.drop("0", axis=1, inplace=True)
    df.rename(columns={'collect_count': 'favorited_count', 'digg_count': 'likes_count', "desc": "description"}, inplace=True)

    column_sums = df.sum()

    # Find columns with a sum equal to 0, empty, or NaN
    columns_to_drop = column_sums[column_sums == 0].index.tolist()
    # These columns
    for i in columns_to_drop:
      if i in cols:
        cols[i] += 1
      else:
        cols[i] = 1

    # Apply the function to each row in the DataFrame
    df['unique_hashtags'] = df['description'].apply(extract_hashtags)

    # Remove the '#' from each hashtag in the list
    df['unique_hashtags'] = df['unique_hashtags'].apply(lambda x: [tag.replace('#', '') for tag in x])

    # Create new column for whether the video is a reply to another video
    pattern = r'^Replying to @\w+'
    df['is_reply'] = df['description'].str.contains(pattern)


    output_path = downloads_folder_path + xlsx_file.replace('final.xlsx', 'cleaned_final.xlsx')
    df.to_excel(output_path, engine='openpyxl', index=False)


xlsx_files = ['Vape_pregnant_cleaned_final.xlsx', 'Vape_pregnancy_cleaned_final.xlsx', 'Vaping_pregnancy_cleaned_final.xlsx', 'e-cigarette_pregnant_cleaned_final.xlsx', 'e-cigarette_pregnancy_cleaned_final.xlsx', 'Vaping_pregnant_cleaned_final.xlsx']
keys = list(cols.keys())
for i in keys:
  if cols[i] < 3:
    # Delete columns that are commonly found to be insignificant (< 3 of 6 files found them to be 0 or empty)
    del cols[i]


column_sets = []
for xlsx_file in xlsx_files:

    xlsx_file_path = downloads_folder_path + xlsx_file

    df = pd.read_excel(xlsx_file_path, engine='openpyxl')
    # For all columns to be removed, make sure this column is present in this Excel file to prevent error
    columns_to_drop_exist = [col for col in columns_to_drop if col in df.columns]

    # Drop all insignificant (0/empty) columns
    df = df.drop(columns=columns_to_drop_exist)
    column_sets.append(set(df.columns))
    df.to_excel(xlsx_file_path, engine='openpyxl', index=False)

# Find the intersection of all sets to get common columns
common_columns = list(set.intersection(*column_sets))

for xlsx_file in xlsx_files:
    xlsx_file_path = downloads_folder_path + xlsx_file
    df = pd.read_excel(xlsx_file_path, engine='openpyxl')
    for c in df.columns:
      if c not in common_columns:
        df.drop(c, axis=1, inplace=True)

    # video id format keeps rounding off last values, so this resets to the full video_id value
    df["video_id"] = df['share_url'].str.extract(r'\/video\/(\d+)\?_r=', expand=False)

    print(xlsx_file, "Num Columns, before:", len(pd.read_excel(xlsx_file_path.replace('_cleaned_final.xlsx', '.xlsx'), engine='openpyxl').columns), "Num Columns, after", len(df.columns))
    df.to_excel(xlsx_file_path, engine='openpyxl', index=False)

print(common_columns)
```

```python
import pandas as pd
import re

xlsx_files = ['Vape_pregnant_cleaned_final.xlsx', 'Vape_pregnancy_cleaned_final.xlsx', 'Vaping_pregnancy_cleaned_final.xlsx', 'e-cigarette_pregnant_cleaned_final.xlsx', 'e-cigarette_pregnancy_cleaned_final.xlsx', 'Vaping_pregnant_cleaned_final.xlsx']
dfs = []
for xlsx_file in xlsx_files:
  downloads_folder_path = '/content/drive/My Drive/'
  xlsx_file_path = downloads_folder_path + xlsx_file

  df = pd.read_excel(xlsx_file_path, engine='openpyxl')
  df["keyword"] = " ".join(xlsx_file.split("_")[:2])
  dfs.append(df)

concatenated_df = pd.concat(dfs, ignore_index=True)
concatenated_df.reset_index(drop=True, inplace=True)

file_name = "vaping_pregnancy_keywords_data.xlsx"
xlsx_file_path = downloads_folder_path + file_name
concatenated_df["video_id"] = concatenated_df['share_url'].str.extract(r'\/video\/(\d+)\?_r=', expand=False)
concatenated_df.to_excel(xlsx_file_path, engine='openpyxl', index=False)

concatenated_df
```
